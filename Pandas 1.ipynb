### Finding min, max and average of a column:

 `min() :` To find minimum of a column <br>
`max() :` To find maximum of a column <br>
`mean():` To find average of a column

```python
#Printing minimum value of Visibility
d['Visibility_km'].min()
```

```python
#Printing miximum value of Visibility
d['Visibility_km'].max()
```

```python
#Printing mean value of Visibility
d['Visibility_km'].mean()
```
